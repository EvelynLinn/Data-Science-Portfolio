```python
import bs4 as bs
import requests
import pandas as pd
```

[Reference](https://codingandfun.com/python-scraping-how-to-get-sp-500-companies-from-wikipedia/)

```python
resp = requests.get('http://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
soup = bs.BeautifulSoup(resp.text, 'lxml')
table = soup.find('table', {'class': 'wikitable sortable'})
print(table)
```

```python
print(table.findAll('tr')[1:])
```

```python
tickers = []
industries = []
sectors = []
headquarters = []
for row in table.findAll('tr')[1:]:
        ticker = row.findAll('td')[0].text
        #fourth element is the sector
        sector = row.findAll('td')[3].text
        industry = row.findAll('td')[4].text
        headquarter = row.findAll('td')[5].text
        
        tickers.append(ticker)
        industries.append(industry)
        sectors.append(sector)
        headquarters.append(headquarter)
```

```python
tickers = list(map(lambda s: s.strip(), tickers))
industries = list(map(lambda s: s.strip(), industries))
sectors = list(map(lambda s: s.strip(), sectors))
headquarters = list(map(lambda s: s.strip(), headquarters))

print(tickers)
print(industries)
```

```python
tickerdf = pd.DataFrame(tickers,columns=['ticker'])
subindustrydf = pd.DataFrame(industries,columns=['Industry'])
sectordf = pd.DataFrame(sectors,columns=['Sector'])
headquarterdf = pd.DataFrame(headquarters,columns=['Headquarter'])

firm_info = pd.concat([tickerdf, sectordf, subindustrydf, headquarterdf], axis=1)
print(firm_info)
```
