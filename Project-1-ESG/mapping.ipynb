```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data/df.csv')
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df.head(20)
```

```python
df[['City', 'State']] = df['Headquarter'].str.split(', ', n=1, expand=True)
df.head()
```

```python
df['State'].unique()
```

```python
countries = ['Ireland', 'Switzerland', 'United Kingdom', 'Netherlands', 'UK', 'Kingdom of the Netherlands']
```

```python
# df['Country'] = np.where(df['State'] is in ? countries, df['State'], 'US')
df['Country'] = np.where(df['State'].isin(countries), df['State'], 'US')
```

```python
df.head()
```

```python
#long_country_names = ['Kingdom of the Netherlands', 'United Kingdom']
df.loc[(df.Country == 'Kingdom of the Netherlands'),'Country']= 'Netherlands'
df.loc[(df.Country == 'United Kingdom'),'Country']= 'UK'
```

```python
df.groupby(['Country']).mean()
```

```python
df.to_csv('Data/mapping.csv')
```
